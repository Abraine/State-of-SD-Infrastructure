```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re 

```

```python
col_name = ['city','county','accdate','holiday','street','alcres','drugres'] #Names of the columns set by us
MVA_DF = pd.DataFrame(data=None, index=None, columns=col_name, dtype=None) #New empty dataframe
size=0
for i in range (9):
    name = 'MVA_' + str(i+2006) #File name
    DF = pd.read_csv('Data_MVA_Fatal/%s.csv' % name,sep = ';', dtype='unicode') #Read the csv file
    r = re.compile(".*city$|.*county$|.*accdate$|.*holiday$|(.*trafid)(1)?$|.*alcres$|.*drugres1$") #Regexp for the col names
    newlist = list(filter(r.match, list(DF.columns))) #Apply regexp to the df columns
    DF = DF[newlist].dropna() #Drop the NaN values
    size += len(DF) #Check size matches 
    DF.columns = col_name #Rename the columns of DF with col_name
    MVA_DF = MVA_DF.append(DF)#Add DF to our initialized dataframe

     
print(len(MVA_DF))#Just to check
print(size)#Just to check
```

```python
#Take away all the cases involved with drugs, alcohol and from SD
MVA_DF['alcres'] = pd.to_numeric(MVA_DF['alcres'],errors='coerce')
MVA_DF['drugres'] = pd.to_numeric(MVA_DF['drugres'],errors='coerce')

MVA_DF = MVA_DF.loc[(MVA_DF['alcres'] <= 8) | (MVA_DF['alcres'] > 95)]
MVA_DF =MVA_DF.loc[(MVA_DF['drugres'] <= 1) | (MVA_DF['drugres'] > 996)]
MVA_DF=MVA_DF.loc[(MVA_DF['county'] == '73')]
MVA_DF=MVA_DF.loc[(MVA_DF['city']== '3260')]

MVA_DF
```
